## Issue Description

Fill missing Region values by look up in the other rows

## Cleaning Code

```python
# Subset of clean dataframe for filtering None values
df_country_region_without_nan = df_clean[df_clean["Region"] != "None"]
```

```python
country_region_dictionary = dict(zip(df_country_region_without_nan.Country, df_country_region_without_nan.Region))
```

```python
country_region_dictionary
```

```python
# Create a function that fills missing Region values
def Region_Lookup(row):
    if(row["Region"] == "None"):
        if (row["Country"] in country_region_dictionary):
            row["Region"] = country_region_dictionary[row["Country"]]
    return row
            
```

```python
df_clean = df_clean.apply(Region_Lookup, axis = 1)
```

# Store

Save the clean dataframes into csv files, then compress them into a zip file. After that delete the old extracted "dirty" csv files.

## File Names

```python
# Zip name
zip_name = zip_name[:-4] + "_clean.zip"
```

```python
# Csv file names
csv_name = "word_happines_clean.csv"
```

## Create CSV Files

```python
# Save the dataframes as csv files
df_clean.to_csv(csv_name, index = False)
```

## Create the Zip File

```python
# compress the csv files
with zipfile.ZipFile(zip_name, "w") as _zip:
    _zip.write(csv_name)
```
